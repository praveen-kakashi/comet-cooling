```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
high_m = []
high_c = []
for i in range(1,13):
    arr2 = np.array(pd.read_csv(r"/home/praveen/Downloads/CoMeT-main/hotspot_tool/oldarch/test3/"+str(i)+"/temperature_mem.trace", sep = "\t", header = None))
    cores = []
    mem = []
    for i in range(4):
        cores.append(arr2[500][i])
    high_c.append(max(cores))
    for i in range(23, 175):
        if arr2[0][i] not in ('X1', 'X2', 'X3'):
            mem.append(arr2[500][i])
    high_m.append(max(mem))
    
xaxis = np.arange(1, 13)
plt.plot(xaxis, high_m, label = "highest memory banks temperature")
plt.plot(xaxis, high_c, label = "highest cores temperature")
plt.xlabel("core power in watts")
plt.ylabel("temperature in Celsius")
plt.legend()
plt.grid()
```
