```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
high_m = []
high_c = []
for i in range(1,10):
    arr2 = np.array(pd.read_csv(r"/home/praveen/Downloads/CoMeT-main/hotspot_tool/oldarch/test"+str(i)+"/3/temperature_mem.trace", sep = "\t", header = None))
    cores = []
    mem = []
    for i in range(4):
        cores.append(arr2[500][i])
    high_c.append(max(cores))
    for i in range(23, 175):
        if arr2[0][i] not in ('X1', 'X2', 'X3'):
            mem.append(arr2[500][i])
    high_m.append(max(mem))
    
xaxis = [0.005, 0.01,0.05,0.1,0.15,0.2,0.25,0.27,0.3]
plt.plot(xaxis, high_m, label = "highest memory banks temperature")
plt.plot(xaxis, high_c, label = "highest cores temperature")
plt.xlabel("bank power in watts")
plt.ylabel("temperature in Celsius")
plt.legend()
plt.grid()
```
