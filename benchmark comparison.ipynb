```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
bms = ['exch', 'egnm', 'gcc', 'lbm', 'gnml']
```

```python
stalls = []
stalls2 = []
for i in range(len(bms)):
    arr = np.array(pd.read_csv(r"/home/praveen/Downloads/CoMeT-main/hotspot_tool/oldarch/t_thresh_"+bms[i]+"_77.trace", sep = "\t", header = None))
    stalls.append(len(arr)-998)
    arr2 = np.array(pd.read_csv(r"/home/praveen/Downloads/CoMeT-main/hotspot_tool3/oldarch/t_thresh_"+bms[i]+"_77.trace", sep = "\t", header = None))
    stalls2.append(len(arr2)-998)
```

```python
X = np.arange(5)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(X + 0.00, stalls, color = 'b', width = 0.25, label = "HMC")
# ax.bar(X + 0.25, stalls2, color = 'g', width = 0.25, label = "HBM")
# ax.legend(labels=['HMC', 'HBM'])
plt.ylabel("number of stalls")
ax.set_xticks(X, ('exch', 'egnm', 'gcc', 'lbm', 'gnml'))
plt.legend
```
