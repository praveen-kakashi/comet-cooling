```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
xaxis = np.arange(0, 17)
highest_b = []
highest_c = []
arr2 = np.array(pd.read_csv(r"/home/praveen/Downloads/CoMeT-main/hotspot_tool/oldarch/test3/3/temperature_mem.trace", sep = "\t", header = None))
cores = []
banks = []
for k in range(23, 175):
    if arr2[0][k] not in('X1', 'X2', 'X3'):
        banks.append(float(arr2[500][k]))
highest_b.append(max(banks))
for k in range(4):
    cores.append(float(arr2[500][k]))
highest_c.append(max(cores))

for s in range(1, 17):
    banks = []
    cores = []
    arr = np.array(pd.read_csv(r"/home/praveen/Downloads/CoMeT-main/hotspot_tool/oldarch/channel"+str(s)+"/test3/3/temperature_mem.trace", sep = "\t", header = None))
    for k in range(23, 175):
        if arr[0][k] not in('X1', 'X2', 'X3'):
            banks.append(float(arr[500][k]))
    highest_b.append(max(banks))

    for k in range(4):
        cores.append(float(arr[500][k]))
    highest_c.append(max(cores))
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(xaxis + 0.00, highest_b, color = 'b', width = 0.50, label = 'highest memory temperature')
# ax.bar(xaxis + 0.250, avgs_c, color = 'g', width = 0.250, label = 'average cores temperature')
plt.xlabel("channel number turned off")
plt.ylabel("temperature in celsius")
plt.xticks(np.arange(0, 17, step=1))
plt.legend()
plt.grid()
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# ax.bar(xaxis + 0.00, avgs_b, color = 'b', width = 0.50, label = 'average memory temperature')
ax.bar(xaxis + 0.00, highest_c, color = 'g', width = 0.50, label = 'highest cores temperature')
plt.xlabel("channel number turned off")
plt.ylabel("temperature in celsius")
plt.xticks(np.arange(0, 17, step=1))
plt.legend()
plt.grid()
plt.show()
```
